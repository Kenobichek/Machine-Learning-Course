```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('lab_1_train.csv', delimiter=',')

id = data[:, 0]
x = data[:, 1]
y = data[:, 2]

plt.scatter(x, y)
plt.title('Train data')
plt.show()
```

```python
def lossFunction(x, y, b0, b1):
    m = len(y)
    y_pred = b0 + b1 * x
    mse = 1 / (2 * m) * sum((y_pred - y) ** 2)
    return mse
```

```python
learning_rate = 0.01
tolerance = 0.0001

b0, b1 = 0, 0
prev_b0, prev_b1 = b0, b1
m = len(y)

while True:
    # predict y value based on current weights
    next_y = b0 + b1 * x
    
    # calculating gradients for each weight
    db0 = (1/m) * sum(next_y - y)
    db1 = (1/m) * sum((next_y - y) * x)
    
    # weight update
    b0 = b0 - learning_rate * db0
    b1 = b1 - learning_rate * db1
    
    # check stop conditions
    if abs(b0 - prev_b0) < tolerance and abs(b1 - prev_b1) < tolerance:
        break
    
    # saving previous weights
    prev_b0, prev_b1 = b0, b1
    
    # loss calculation function
    mse = lossFunction(x, y, b0, b1)
    print("MSE:", mse)

print("Coefficients b0 and b1: ", b0, b1)
```

```python
test_data = np.loadtxt('lab_1_test.csv', delimiter=',')

test_id = test_data[:, 0]
test_x = test_data[:, 1]
test_y = test_data[:, 2]

plt.scatter(test_x, test_y, c='#ff00ff')
plt.title('Test data')
plt.show()
```

```python
test_mse = lossFunction(test_x, test_y, b0, b1)

print("MSE on test data:", test_mse)
```
